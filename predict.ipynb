```python
import pandas as pd
# Convert to DataFrame
data = pd.read_excel('/home/mohammad/Downloads/test.xlsx')

# Convert columns to required types
data["date_listed"] = pd.to_datetime(data["date_listed"])  # Convert to datetime
data["year"] = data["year"].astype("int16")               # Convert to int16
data["manufacturer"] = data["manufacturer"].astype("object")  # Convert to categorical
data["model"] = data["model"].astype("object")
data["type"] = data["type"].astype("object")
data["connectivity"] = data["connectivity"].astype("object")
data["hours"] = data["hours"].astype("object")
data["brand"] = data["brand"].astype("object")
data["country"] = data["country"].astype("object")
data["condition"] = data["condition"].astype("object")
data["accessories"] = data["accessories"].astype("int8")   # Convert to int8
data["discount_percent"] = data["discount_percent"].astype("int8")
data["user_rating"] = data["user_rating"].astype("float32")  # Convert to float32
data["warranty_extension_available"] = data["warranty_extension_available"].astype("int8")

```

```python
data = data.drop(columns=['device_id'])
data.head()
```
